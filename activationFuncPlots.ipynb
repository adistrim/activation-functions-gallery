```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def prelu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def swish(x):
    return x * sigmoid(x)

def hard_sigmoid(x):
    return np.clip((x + 1) / 2, 0, 1)

def softplus(x):
    return np.log1p(np.exp(x))

def softsign(x):
    return x / (1 + np.abs(x))
```

```python
x = np.linspace(-10, 10, 400)

sigmoid_y = sigmoid(x)
relu_y = relu(x)
tanh_y = tanh(x)
leaky_relu_y = leaky_relu(x)
prelu_y = prelu(x)
elu_y = elu(x)
swish_y = swish(x)
hard_sigmoid_y = hard_sigmoid(x)
softplus_y = softplus(x)
softsign_y = softsign(x)
```

```python
plt.figure(figsize=(20, 10))
plt.subplot(2, 2, 1)
plt.plot(x, sigmoid_y, label='Sigmoid Function', color='blue')
plt.title('Sigmoid Activation Function')
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.subplot(2, 2, 2)
plt.plot(x, relu_y, label='ReLU Function', color='blue')
plt.title('ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.subplot(2, 2, 3)
plt.plot(x, tanh_y, label='Tanh Function', color='blue')
plt.title('Tanh Activation Function')
plt.xlabel('x')
plt.ylabel('Tanh(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.subplot(2, 2, 1)
plt.plot(x, leaky_relu_y, label='Leaky ReLU Function', color='blue')
plt.title('Leaky ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('Leaky ReLU(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.subplot(2, 2, 1)
plt.plot(x, prelu_y, label='Parametric ReLU Function', color='blue')
plt.title('Parametric ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('Parametric ReLU(x)')
plt.grid(True)
plt.legend()
plt.show()
```
