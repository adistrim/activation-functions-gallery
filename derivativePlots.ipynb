```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)

def prelu_derivative(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)

def elu_derivative(x, alpha=1.0):
    return np.where(x > 0, 1, alpha * np.exp(x))

def swish_derivative(x):
    sig = sigmoid(x)
    return sig + x * sig * (1 - sig)

def hard_sigmoid_derivative(x):
    return np.where((x >= -1) & (x <= 1), 1/2, 0)

def softplus_derivative(x):
    return sigmoid(x)

def softsign_derivative(x):
    return 1 / (1 + np.abs(x))**2
```

```python
x = np.linspace(-10, 10, 400)

sigmoid_derivative_y = sigmoid_derivative(x)
relu_derivative_y = relu_derivative(x)
tanh_derivative_y = tanh_derivative(x)
leaky_relu_derivative_y = leaky_relu_derivative(x)
prelu_derivative_y = prelu_derivative(x)
elu_derivative_y = elu_derivative(x)
swish_derivative_y = swish_derivative(x)
hard_sigmoid_derivative_y = hard_sigmoid_derivative(x)
softplus_derivative_y = softplus_derivative(x)
softsign_derivative_y = softsign_derivative(x)
```

```python
plt.figure(figsize=(20, 15))
plt.subplot(3, 2, 2)
plt.plot(x, sigmoid_derivative_y, label='Sigmoid Derivative', color='blue')
plt.title('Sigmoid Derivative')
plt.xlabel('x')
plt.ylabel('Sigmoid Derivative(x)')
plt.grid(True)
plt.legend()
plt.show()
```
